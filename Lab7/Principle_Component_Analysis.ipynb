```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([[2.5,3.2], [1.3,4.2], [2.8,2.7], [2.5,3.2], [3.5,4.7], [3.0,4.5], [2.8,3.8], [3.3,4.5], [2.4,4.0], [2.5,4.2], [2.9,4.0], [2.7,4.1], [2.8,4.5], [3.0,4.2], [2.8,4.3], [3.0,4.3], [2.9,4.4], [3.0,4.5], [3.1,4.7]])
```

```python
plt.scatter(data[:,0], data[:,1])
```

```python
mean_vector = np.mean(data, axis=0)
print(mean_vector)
```

```python
centered_data = data - mean_vector
print(centered_data)
```

```python
plt.scatter(centered_data[:,0], centered_data[:,1])
```

```python
cov_matrix = np.cov(centered_data, rowvar=False)
print(cov_matrix)
```

```python
np.linalg.matrix_rank(cov_matrix)
```

```python
eigenValues, eigenVectors = np.linalg.eig(cov_matrix)
print("eigenValues = ",eigenValues)
print("eigenVectors : \n",eigenVectors)
```

```python
# nplinalg.norm(x,axis = ) #to normalize the vector
```

```python
principle_component = eigenVectors[:,[1,0]]
print(principle_component)
```

```python
DV = np.dot(centered_data, principle_component)
print(DV)   
```

```python
plt.quiver(0,0,[1],[0],color='r',label='PC1', scale=1)  
plt.quiver(0,0,[0],[1],color='b',label='PC2', scale=1)
plt.grid()
plt.scatter(DV[:,0], DV[:,1]) 
plt.legend()
```

```python
plt.quiver(0,0,*principle_component[:,0],color='r',label='PC1', scale=0.5)  
plt.quiver(0,0,*principle_component[:,1],color='b',label='PC2', scale=0.5)
plt.grid()
plt.scatter(centered_data[:,0], centered_data[:,1]) 
plt.legend()
```
