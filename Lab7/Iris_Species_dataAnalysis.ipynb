```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt

iris = load_iris()
data = iris.data
target = iris.target
target_names = iris.target_names
print(data)
```

```python
mean_vector = np.mean(data, axis=0)
print(mean_vector)
```

```python
centered_data = data - mean_vector
print(centered_data)
```

```python
cov_matrix = np.cov(centered_data, rowvar=False)
print(cov_matrix)
```

```python
eigenValues, eigenVectors = np.linalg.eig(cov_matrix)
print("eigenValues = ",eigenValues)
print("eigenVectors : \n",eigenVectors)
```

```python
DV = np.dot(centered_data, eigenVectors)
print(DV)
```

```python
reduced_data = DV[:,:2]
print(reduced_data)
```

```python
plt.xlim(-4,4)
plt.ylim(-4,4)
# plt.quiver(0,0,[1],[0],color='r',label='PC1', scale=1)  
# plt.quiver(0,0,[0],[1],color='b',label='PC2', scale=1)
plt.grid()
for label, color, marker in zip(range(3), ['r','g','b'], ['o','^','s']):
    # plt.scatter(reduced_data[tar1get==label,0], reduced_data[target==label,1], c=color, label=target_names[label], marker=marker)
    plt.scatter(DV[target == label,0], DV[target == label,1], label=target_names[label], c=color, marker=marker) 
plt.legend()
```
