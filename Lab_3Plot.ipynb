```python
import matplotlib.pyplot as pit
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = 4 + 1*np.sin(0.9 * x)
```

```python
print(f"x = {x}, \n\ny = {y}")
```

```python
pit.plot(x,y)
```

```python
pit.plot(x,y,color='blue', linewidth = 2, marker='.')
```

```python
pit.plot(x,y,linewidth = 2, color = 'red')
#ADD labels, title and grid
pit.title('Line Plot')
pit.xlabel('X')
pit.ylabel('Y')
```
